```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV
df = pd.read_csv('../volta_simulated_data.csv', parse_dates=['timestamp'])

# Preview data
df.head()
```

```python
home_1 = df[df['home_id'] == 1]

plt.figure(figsize=(15, 6))
plt.plot(home_1['timestamp'], home_1['solar_generation_kw'], label='Solar Generation (kW)')
plt.plot(home_1['timestamp'], home_1['grid_demand_kw'], label='Grid Demand (kW)')
plt.plot(home_1['timestamp'], home_1['battery_level_pct'], label='Battery Level (%)')
plt.plot(home_1['timestamp'], home_1['grid_draw_kw'], label='Grid Draw (kW)')
plt.title("Energy Behavior for Home 1")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
